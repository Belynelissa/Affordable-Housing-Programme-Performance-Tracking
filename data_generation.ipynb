# Project Scenario Design

To ensure the synthetic dataset reflects real-world project behaviour, project timelines were designed around five business scenarios commonly observed in project portfolio management.

The scenarios are:

1. **Ongoing - On Track**
   - The project is still in progress and remains within its original timeline.

2. **Delayed**
   - The project has exceeded its planned completion date and remains incomplete.

3. **Ongoing - Extension Granted**
   - The original completion date has passed, an extension was approved, and the project is still in progress.

4. **Completed - On Time**
   - The project was completed on or before its planned completion date without requiring an extension.

5. **Completed - After Extension**
   - The project exceeded its original timeline, received an extension, and was completed within the revised schedule.

These scenarios were incorporated into the data generation process to create a realistic project portfolio and support meaningful performance analysis in Power BI.

# Dataset Generation Logic

After defining the project scenarios and business rules, the next step is to generate a synthetic project portfolio.

The dataset generation process follows the logic below:

1. Randomly select a project scenario based on predefined probabilities.
2. Generate project dates according to the business rules of that scenario.
3. Generate project progress metrics such as completion percentage and units/stalls completed.
4. Generate financial information, including allocated and spent budgets.
5. Assign project attributes such as county, contractor, and project name.
6. Store the generated project record.
7. Repeat the process until the required number of projects has been created.

## Generation Workflow

```text
Select Scenario
        ↓
Generate Dates
        ↓
Generate Progress Metrics
        ↓
Generate Financial Metrics
        ↓
Assign Project Attributes
        ↓
Store Project Record
        ↓
Repeat Until Dataset Is Complete
```

This rule-based approach ensures that the synthetic dataset remains internally consistent and closely mimics the behaviour of real project portfolios while protecting sensitive information.

````python
# Reference data
counties = [
    "Nairobi", "Kiambu", "Kisumu",
    "Mombasa", "Nakuru", "Uasin Gishu"
]

contractors = [
    "BuildKen Ltd",
    "AfriConstruct",
    "Homeland Developers",
    "UrbanRise Contractors",
    "Skyline Builders"
]
````

````python
#Scenarios and probabilities
scenarios = [
    "ongoing_on_track",
    "delayed",
    "ongoing_extension",
    "completed_on_time",
    "completed_after_extension"
]

probabilities = [0.35, 0.20, 0.15, 0.20, 0.10]
````

````python
#Housing dataset
n_projects = 250
#Storage container
housing_data = []

for i in range(1, n_projects + 1):

    scenario = np.random.choice(scenarios, p=probabilities)

    county = random.choice(counties)
    contractor_id = random.randint(1, len(contractors))

    start_date = today - timedelta(days=random.randint(200, 1200))
    planned_end_date = start_date + timedelta(days=random.randint(180, 800))

    revised_end_date = None
    actual_end_date = None
    status = ""
    progress = 0

    if scenario == "ongoing_on_track":
        status = "Ongoing"
        progress = random.randint(10, 90)

    elif scenario == "delayed":
        status = "Delayed"
        progress = random.randint(10, 95)

    elif scenario == "ongoing_extension":
        status = "Ongoing"
        revised_end_date = planned_end_date + timedelta(days=random.randint(60, 300))
        progress = random.randint(40, 95)

    elif scenario == "completed_on_time":
        status = "Completed"
        actual_end_date = planned_end_date - timedelta(days=random.randint(0, 30))
        progress = 100

    elif scenario == "completed_after_extension":
        status = "Completed"
        revised_end_date = planned_end_date + timedelta(days=random.randint(60, 300))
        actual_end_date = revised_end_date - timedelta(days=random.randint(0, 30))
        progress = 100

    planned_units = random.randint(80, 600)
    completed_units = int(planned_units * (progress / 100))

    budget_allocated = random.randint(20_000_000, 120_000_000)
    budget_spent = budget_allocated * random.uniform(0.6, 1.3)

    # ✅ THIS MUST BE INSIDE THE LOOP
    housing_data.append([
        i,
        f"Housing_Project_{i}",
        county,
        random.randint(1, 5),
        contractor_id,
        start_date,
        planned_end_date,
        revised_end_date,
        actual_end_date,
        planned_units,
        completed_units,
        progress,
        budget_allocated,
        round(budget_spent, 2),
        status,
        scenario
    ])
````

#### Modern Markets

````python
#Set up container
market_data = []
n_markets = 100

#Loop Structure
for i in range(1, n_markets + 1):

    scenario = np.random.choice(scenarios, p=probabilities)

    county = random.choice(counties)
    contractor_id = random.randint(1, len(contractors))

    start_date = today - timedelta(days=random.randint(200, 1200))

    planned_end_date = start_date + timedelta(days=random.randint(180, 900))

    revised_end_date = None
    actual_end_date = None
    status = ""
    progress = 0

    #Scenario logic
    #Ongoing on track
    if scenario == "ongoing_on_track":
        status = "Ongoing"
        progress = random.randint(10, 90)
        #Delayed
    elif scenario == "delayed":
        status = "Delayed"
        progress = random.randint(10, 95)
        #Ongoing extension
    elif scenario == "ongoing_extension":
        status = "Ongoing"
        revised_end_date = planned_end_date + timedelta(days=random.randint(60, 300))
        progress = random.randint(40, 95)
        #Completed on time
    elif scenario == "completed_on_time":
        status = "Completed"
        actual_end_date = planned_end_date - timedelta(days=random.randint(0, 30))
        progress = 100
        #Completed after extension
    elif scenario == "completed_after_extension":
        status = "Completed"
        revised_end_date = planned_end_date + timedelta(days=random.randint(60, 300))
        actual_end_date = revised_end_date - timedelta(days=random.randint(0, 30))
        progress = 100

        #Market-specific metrics
    planned_stalls = random.randint(20, 200)
    completed_stalls = int(planned_stalls * (progress / 100))
     #Budget logic
    budget_allocated = random.randint(50_000_000, 300_000_000)
    budget_spent = budget_allocated * random.uniform(0.6, 1.35)


    #Store record
    market_data.append([
        i,
        f"Modern_Market_{i}",
        county,
        random.randint(1, 5),
        contractor_id,
        start_date,
        planned_end_date,
        revised_end_date,
        actual_end_date,
        planned_stalls,
        completed_stalls,
        progress,
        budget_allocated,
        round(budget_spent, 2),
        status,
        scenario
    ])
````

##### Data quality validation

Before exporting the synthetic datasets, several validation checks were performed to ensure data integrity and consistency.

The checks included:

- Verifying dataset dimensions
- Checking for unexpected missing values
- Ensuring completed projects have actual completion dates
- Ensuring ongoing and delayed projects do not have actual completion dates
- Verifying completed outputs do not exceed planned outputs
- Checking for negative budget values
- Reviewing progress percentage distributions
- Validating project scenario distributions against the predefined probabilities

These checks improve the reliability of the synthetic data and ensure that the generated datasets realistically mimic project portfolio behaviour.
